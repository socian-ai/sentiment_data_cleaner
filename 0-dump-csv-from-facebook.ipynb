```python
import json
import re
import pandas as pd
import glob
import os
from tqdm import tqdm
```

```python
input_file_path = 'input/elasticsearch/combined.json'
output_file_path = 'output/0-facebook.csv'
!mkdir -p output
```

```python
BANGLISH_REGEX = re.compile(r'[A-Za-z0-9]+')

def is_banglish(text):
    if BANGLISH_REGEX.match(text):
        return True
    else:
        return False
```

```python
text_list = []
sources = {}
with open(input_file_path) as input_f:
    for line in tqdm(input_f):
        json_parsed = json.loads(line)
        
        source_id = json_parsed['_source']['source_page_id']
        fb_id = json_parsed['_source']['fb_id']
        text = json_parsed['_source']['message']
        sources[source_id] = True
        
        if not isinstance(text, str):
            continue
            
        if is_banglish(text):
#             print(text)
            continue
        
        if len(text) < 5:
            continue
        
        text_list.append({'text': text, 'source': 'facebook', 'fb_id': fb_id, 'page_id': source_id})

print('Loaded lines: ', len(text_list))
```
